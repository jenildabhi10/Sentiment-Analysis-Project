```python
import pandas as pd
from transformers import pipeline
```

```python
# Load sentiment analysis model
sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
# Load your CSV file
df = pd.read_csv("sentiment_analysis.csv")
```

```python
# Check which column contains text data
print("Columns in your file:", df.columns)
```

```python
# Change this to match your actual text column name
text_column = "text"
```

```python
# Create an empty list to store sentiment labels
sentiments = []
```

```python
# Loop through each text entry and analyze sentiment
for text in df[text_column]:
    result = sentiment_pipeline(str(text))[0]   # get the first result
    sentiments.append(result['label'])          # store only the label (e.g., POSITIVE/NEGATIVE)
```

```python
# Add the new column to the DataFrame
df['sentiment'] = sentiments

# Save the updated CSV file
output_path = "sentiment_analysis_with_sentiment.csv"
df.to_csv(output_path, index=False)

print(f"Sentiment analysis complete! File saved as: {output_path}")
```
